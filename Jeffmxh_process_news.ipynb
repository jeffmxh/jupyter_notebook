```python
import pandas as pd
import re

def load_corpus(file_path):
    print('开始读取')
    raw_data = pd.read_table(file_path, header = None)
    print('读取结束')
    raw_data.columns = ['content']
    pattern = re.compile('<content>.+?</content>')
    filter_pattern = re.compile('\\u3000|\\ue40c|</?content>')
    content_list = []
    for text in raw_data.content:
        if pattern.match(text):
            if len(content_list)%1000==0:
                print('content_list length: ', len(content_list))
            text = filter_pattern.sub('', text)
            if text not in content_list:
                content_list.append(text)
    return content_list
```

```python
def strQ2B(ustring):
    """全角转半角"""
    rstring = ""
    for uchar in ustring:
        inside_code=ord(uchar)
        if inside_code == 12288:                              #全角空格直接转换            
            inside_code = 32 
        elif (inside_code >= 65281 and inside_code <= 65374): #全角字符（除空格）根据关系转化
            inside_code -= 65248

        rstring += chr(inside_code)
    return rstring
```

```python
def preprocess_corpus(text):
    filter_pattern = re.compile('组图\W?|图文\W?|\\(.+?\\)|\[.+?\]|【.+?】|「|，|」|￥|…')
    name_pattern = re.compile('《.+?》')
    url_pattern = re.compile('(https?|ftp|file)://[-A-Za-z0-9+&@#/%?=~_|!:,.;]+[-A-Za-z0-9+&@#/%=~_|]')
    date_pattern = re.compile('\d+年\d+月\d+日|\d+年\d+月|\d+月\d+日|\d+月\d+号|\d+年|\d+月份?|\d+日|\d+小时|\d+分钟|\d\d:\d\d')
    number_pattern = re.compile('\d+')
    text = filter_pattern.sub(',', text)
    text = name_pattern.sub('TAG_NAME', text)
    text = url_pattern.sub('TAG_URL', text)
    text = date_pattern.sub('TAG_DATE', text)
    text = number_pattern.sub('TAG_NUMBER', text)
    return text
```

```python
import os
import re
from os import path
import jieba

class jieba4null():
    """docstring for parser_word
    deal处理文本，返回词表、词性及依存关系三个值"""
    def __init__(self,n_core = 16):
        jieba.load_userdict('/home/jeffmxh/user_dict.txt')
        self.n_CORE=n_core
        jieba.enable_parallel(self.n_CORE-1)
    def cut_word(self,sent):
        words = jieba.cut(sent, cut_all=False)
        result = list(words)
        return ' '.join(result)
```
