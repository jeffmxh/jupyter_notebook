```python
import os
import re
import pandas as pd
from collections import defaultdict
```

```python
'''
用于过滤异常编码字符的类
'''

class char_filter():
    def __init__(self):
        self.filter_pattern = re.compile(r'\\x[a-z0-9]{2}')
        self.filter_title = re.compile("^b'|\{[0-9]+\}")
    def filter(self, text):
        foo = self.filter_pattern.sub('', text)
        return self.filter_title.sub('', foo)
```

```python
'''
根据输入路径读取txt文件，返回一个列表
'''

def load_txt(file_path):
    article = []
    illegal_filter = char_filter()
    with open(file_path, 'rb') as f:
        for line in f:
            foo = str(line)
            article.append(illegal_filter.filter(foo))
    return article
```

```python
'''
定义输入文件路径并读取文件
'''

filePath = '/home/da/task.txt'
article = load_txt(filePath)
article
```

```python
'''
按空格把句子分割为单词
'''

regex = re.compile('\s+')
words = regex.split(article[0])
```

```python
'''
对单词进行计数
'''

counts = defaultdict(lambda: 0)  # 使用lambda来定义简单的函数

for word in words:
    counts[word] += 1
```

```python
'''
生成结果数据框
'''

result = pd.DataFrame({'word':list(counts.keys()), 'value':list(counts.values())})
result.sort_values('value', ascending=False)
```
